QUES-5->Explain the difference between justify-items and justify-self using code examples.
ANS=>JUSTIFY CONTENT  ==>.it is applied to entire container.
                    .scope is global.
                    .it purpose is to set horizontal of all items.
<!-- code is below-->
 <style>
    .container{
        display:grid;
        grid-template-columns: repeat(3,100px);
        justify-items: center;
        gap:10px;
        background:#eee ;
    }
    .box{
        background: #88f;
        color:white;
        padding:10px;
    }
 </style>

JUSTIFY SELF  ==>.it is applied to individual grid items.
                 .scope is local.
                 .purpose is to set the horizontalalignment of specific items.     
                 
                 <!--code is below-->
                 <style>
  .container {
    display: grid;
    grid-template-columns: repeat(3, 100px);
    gap: 10px;
    background: #eee;
  }
  .box {
    background: #88f;
    color: white;
    padding: 10px;
  }
  .box1 {
    justify-self: start;  /* aligns left */
  }
  .box2 {
    justify-self: center; /* aligns center */
  }
  .box3 {
    justify-self: end;    /* aligns right */
  }
</style>
